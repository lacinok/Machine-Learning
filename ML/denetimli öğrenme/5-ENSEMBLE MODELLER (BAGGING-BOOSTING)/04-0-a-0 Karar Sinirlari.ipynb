# 0. Interaktif hucre

Interaktif hucreler, parametrelerin etkilerini gozlemlemek ve diger pek cok amac icin kullanilabilir. 

Asagidaki ornekte bir dogrunun degisen egim parametresine gore hareketi gosterilmektedir (konuyla alakasiz bir ornek):

```python
@interact
def dogruviz(egim=(-2,2,.1)):
    plt.scatter([-1, 1, -1, 1], [1, 1, -1, -1])
    plt.scatter(0, 0)
    plt.axline((0,0), slope=egim, color='red')
    plt.title('Eğim: %.1f' % egim)
    plt.grid()
```
